```python
from skimage import io
image = io.imread("Lena.jpg")
```

```python
io.imshow(image)
image.shape
```

```python
import matplotlib.pylab as plt
plt.hist(image.flatten())
```

```python
from skimage import color

imageGray = color.rgb2gray(image)
io.imshow(imageGray)
```

```python
plt.contour(imageGray, origin = "image")  #origin = "image", else inverted
```

```python
from skimage.transform import rescale
image_rescaled = rescale(imageGray, 0.25)
io.imshow(image_rescaled)
```

```python
from skimage import filters
edges = filters.sobel(imageGray)
io.imshow(edges)
io.show()
```

```python
from skimage import data

camera = data.camera()
print(camera)
io.imshow(camera)
```

```python
camera[:30,:] = 0
camera[:,:30] = 0

plt.imshow(camera, cmap = 'gray')
plt.show()
```

```python
mask = camera < 87
camera[mask] = 255

io.imshow(camera)
```

```python
from scipy import misc, ndimage
   
blurred_face = ndimage.gaussian_filter(image, sigma=3)  
very_blurred = ndimage.gaussian_filter(image, sigma=5)
   
io.imshow(blurred_face)
```

```python
import cv2
import numpy as np

#color
img1 = cv2.imread("Lena.jpg")

cv2.imshow('image',img1)
cv2.waitKey(0)
```

```python
img2 = cv2.imread('Lena.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imshow('image',img2)
cv2.waitKey(0)

cv2.imwrite('Lena1.png',img1)
cv2.imwrite('Lena1.jpg',img2)
```

```python
cap = cv2.VideoCapture(0)
 
while(True):
    ret, frame = cap.read()
    color = cv2.cvtColor(frame, cv2.COLOR_BGR2BGRA)
 
    cv2.imshow('frame', color)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    out.write(frame)
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
