```python
import pandas as pd
```

```python
df = pd.read_csv("Equity Financial Data.csv")
```

```python
df = df.loc[:,["Class","Dividend per Share","10Y Net Income Growth (per Share)","10Y Shareholders Equity Growth (per Share)"]]
```

```python
df = df.dropna()
```

```python
X = df.loc[:,["Dividend per Share","10Y Net Income Growth (per Share)","10Y Shareholders Equity Growth (per Share)"]]
```

```python
Y = df.loc[:, ["Class"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(ccp_alpha=0.00244)#min_samples_split=200 ccp_alpha=0.00244 max_depth=4
```

```python
model.fit(X_train, Y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(Y_test, pred)
```

```python
print((cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X_train, Y_train)
```

```python
model = tree.DecisionTreeClassifier()
path = model.cost_complexity_pruning_path(X_train1,Y_train1)
```

```python
for i in path.ccp_alphas:
    model = tree.DecisionTreeClassifier(ccp_alpha=i)
    model.fit(X_train1,Y_train1)
    pred = model.predict(X_test1)
    cm = confusion_matrix(Y_test1,pred)
    print(i)
    print((cm[0,0]+cm[1,1])/(sum(sum(cm))))
    print("======================")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
g = GridSearchCV(estimator=model, param_grid=dict(ccp_alpha=[i for i in path.ccp_alphas]))
```
